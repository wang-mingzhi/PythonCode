# 画图实例

## 导入第三方库

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

## 多个图形画在一个图表

```python
Z = np.random.rand(2, 11)
x = np.arange(-0.5, 11, 1)  # len = 11
y = np.arange(0, 2, 1)  # len = 7
fig, ax = plt.subplots()
c = ax.pcolor(x, y, Z, cmap="bwr", vmin=Z.min(), vmax=Z.max(), shading='auto')
ax.plot(np.linspace(0, 10, 500), abs(np.sin(np.linspace(0, 10, 500))), color='white', linewidth=3)
fig.colorbar(c, ax=ax)
plt.show()
```
