```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
Z = np.random.rand(2, 11)
x = np.arange(-0.5, 11, 1)  # len = 11
y = np.arange(0, 2, 1)  # len = 7
fig, ax = plt.subplots()
c = ax.pcolor(x, y, Z, cmap="Wistia", vmin=Z.min(), vmax=Z.max(), shading='auto')
ax.plot(np.linspace(0, 10, 500), abs(np.sin(np.linspace(0, 10, 500))), label='Using set_dashes()')
fig.colorbar(c, ax=ax)
plt.show()
```
